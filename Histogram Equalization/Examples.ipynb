```python
"""
Histogram Equalization (Gray)

@author: Rodrigo Schardong

"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

from Histogram_Class import Histogram_Functions
hist = Histogram_Functions()

name = "Lena_c.png" #The image file name
image = cv2.imread(name, 0) #Var image has your input image


out = hist.Histogram_Equalization(image)


#PLOT
f,a = plt.subplots(ncols = 2, figsize=(20,5))
a[0].imshow(image, cmap="gray")
a[0].set_title("Input (Gray)")
a[1].imshow(out.astype(int), cmap="gray")
a[1].set_title("Output (Gray)")
plt.show()
```
